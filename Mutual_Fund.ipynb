```python
import pandas as pd
import numpy as np
from mftool import Mftool
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
```

```python
tickers=['0P0000XUYI.BO','0P0000XVK5.BO','0P0000NRF6.BO']
data=yf.download(tickers, period='2y', interval='1d')
```

```python
data
```

```python
adj_close_df=data['Adj Close']
```

```python
adj_close_df
```

```python
col=list(adj_close_df.columns)
```

```python
col
```

```python
for i in col:
    adj_close_df[i+'%Return']=adj_close_df[i]/adj_close_df[i].shift(1)-1
```

```python
adj_close_df
```

```python
adj_close_df.dropna(inplace=True)
```

```python
adj_close_df
```

```python
adj_close_df2=adj_close_df.iloc[:235, :]
```

```python
adj_close_df2
```

```python
return_columns=list(adj_close_df.columns)[-3:]
```

```python
return_columns
```

```python
return1=[]
risk1=[]
for i in return_columns:
    temp=adj_close_df2[i].mean()
    temp2=((temp+1)**250-1)*100
    return1.append(temp2)
    temp3=adj_close_df2[i].std()*np.sqrt(250)*100
    risk1.append(temp3)
```

```python
risk_return_df=pd.DataFrame(index=col)
risk_return_df['Risk%']=risk1
risk_return_df['Return%']=return1
risk_return_df['Risk Adj Return']=risk_return_df['Return%']/risk_return_df['Risk%']
```

```python
risk_return_df
```

```python
risk_return_list=list(risk_return_df['Risk Adj Return'])
sum_list=np.sum(risk_return_list)
weight_array=risk_return_list/sum_list
risk_return_df['Weightage']=weight_array
```

```python
risk_return_df
```

```python
return_df=adj_close_df.iloc[:,-3:]
```

```python
return_df
```

```python
return_df=return_df.iloc[-235:,:]
```

```python
return_df
```

```python
rebased_df=pd.DataFrame()
for j in list(return_df.columns):
    rebased_price_list=[100]
    for i in list(return_df[j]):
        temp=rebased_price_list[-1]*(1+i)
        rebased_price_list.append(temp)
    rebased_df[j]=rebased_price_list
```

```python
rebased_df
```

```python
rebased_df.rename(columns={'0P0000NRF6.BO%Return':'ICICI Prud RegSave Direct','0P0000XUYI.BO%Return':'SBI Magnum Global Direct','0P0000XVK5.BO%Return':'SBI Life Index'},inplace=True)
rebased_df
```

```python
final_df=pd.DataFrame()
final_df['Portfolio NAV']=rebased_df.dot(list(risk_return_df['Weightage']))
```

```python
final_df
```

```python
ticker='^NSEI'
nsei=yf.download(ticker, period='1y',interval='1d')
```

```python
nsei['NSEI % Return']= nsei['Adj Close']/nsei['Adj Close'].shift(1)-1
nsei.dropna(inplace=True)
rebased_price_list1=[100]
for i in list(nsei['NSEI % Return']):
    temp=rebased_price_list1[-1]*(i+1)
    rebased_price_list1.append(temp)
```

```python
final_df['NIFTY']=rebased_price_list1[:-10]
```

```python
final_df
```
